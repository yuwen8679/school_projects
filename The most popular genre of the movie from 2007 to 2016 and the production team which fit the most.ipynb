## 3 & 4. Subsets of Data & Visualization

### (1) The subset of genre

##### Before we find the director, we need to find the type of movie which has the highest revenue in the past 10 years in order to know which kind of movie people prefer.

```python
#Select the data which is from 2007 to 2016
data1=data.loc[(data["year"] ==2007)| (data["year"] == 2008) | (data["year"] == 2009)| (data["year"] == 2010) | (data["year"] == 2011) | (data["year"] == 2012)
             | (data["year"] == 2013)| (data["year"] == 2014)| (data["year"] == 2015)| (data["year"] == 2016)]

#Use group by function to get the total gross of each genre
datab=data1.groupby('genre',as_index=False).sum()

#Sort the data by gross
datab=datab.sort_values(by=['gross'],ascending=False)

##Select the columns of genre and gross
datab=datab[["genre","gross"]]

#Print out the result
datab
```

##### Action movies seems to have the largest amount of revenue from 2007 to 2016

##### In order to see the state of the past 10 years, we would pick 6 of the genre which have the largest gross then calculate the revenue of each genre by year.

```python
#Select all of the Action movies from 2007 to 2016
Action=data1.loc[(data1["genre"] =="Action")]

#Select the columns of genre,gross,and year
Action=Action[['genre',"gross","year"]]

#Sum up the data of each year
Action=Action.groupby('year',as_index=False).sum()


#Select all of the Comedy movies from 2007 to 2016
Comedy=data1.loc[(data1["genre"] =="Comedy")]

#Select the columns of genre,gross,and year
Comedy=Comedy[['genre',"gross","year"]]

#Sum up the data of each year
Comedy=Comedy.groupby('year',as_index=False).sum()

#Select all of the Animation movies from 2007 to 2016
Animation=data1.loc[(data1["genre"] =="Animation")]

#Select the columns of genre,gross,and year
Animation=Animation[['genre',"gross","year"]]

#Sum up the data of each year
Animation=Animation.groupby('year',as_index=False).sum()

#Select all of the Adventure movies from 2007 to 2016
Adventure=data1.loc[(data1["genre"] =="Adventure")]

#Select the columns of genre,gross,and year
Adventure=Adventure[['genre',"gross","year"]]

#Sum up the data of each year
Adventure=Adventure.groupby('year',as_index=False).sum()

#Select all of the Drama movies from 2007 to 2016
Drama=data1.loc[(data1["genre"] =="Drama")]

#Select the columns of genre,gross,and year
Drama=Drama[['genre',"gross","year"]]

#Sum up the data of each year
Drama=Drama.groupby('year',as_index=False).sum()

#Select all of the Biography movies from 2007 to 2016
Biography=data1.loc[(data1["genre"] =="Biography")]

#Select the columns of genre,gross,and year
Biography=Biography[['genre',"gross","year"]]

#Sum up the data of each year
Biography=Biography.groupby('year',as_index=False).sum()

#Combine all the genre data
frames = [Action["gross"], Comedy["gross"],Animation["gross"],Adventure["gross"],Drama["gross"],Biography["gross"]]

#Combine all the genre data
result = pd.concat(frames,axis=1, join='outer', keys=['Action', 'Comedy','Animation','Adventure','Drama','Biography'])

#Add the row names of the data frame which is the years.
result.index=['2007','2008','2009','2010','2011','2012','2013','2014','2015','2016']

#Print out the result
result
```

##### We could not easy to find the best genres by the table, but it would be much clear on the plot.


### (1) The visualization of genre

##### Draw the line chart of each genre in order to see the state of the past 10 years

```python
import matplotlib.pyplot as plt

from matplotlib.font_manager import FontProperties

#Set the plot size
plt.figure(figsize=(15,10),dpi=100,linewidth = 2)

#Use Year as X-axis data
year = Action['year']

#Select the column of gross from Action data 
Ac=Action['gross']

#Put the line on the plot which represent the annual revenue of Action movies
plt.plot(year,Ac,'s-',color = 'r', label="Action")

#Select the column of gross from Adventure data 
Adv=Adventure['gross']

#Put the line on the plot which represent the annual revenue of Adventure
plt.plot(year,Adv,'s-',color = 'b', label="Adventure")

#Select the column of gross from Animation data
Ani=Animation['gross']

#Put the line on the plot which represent the annual revenue of Animation
plt.plot(year,Ani,'s-',color = 'g', label="Animation")

#Select the column of gross from Biography data
bio=Biography['gross']

#Put the line on the plot which represent the annual revenue of Biography
plt.plot(year,bio,'s-',color = 'lightblue', label="Biography")

#Select the column of gross from Comedy data
com=Comedy['gross']

#Put the line on the plot which represent the annual revenue of Comedy
plt.plot(year,com,'s-',color = 'pink', label="Comedy")

#Select the column of gross from Drama data
dra=Drama['gross']

#Put the line on the plot which represent the annual revenue of Drama
plt.plot(year,dra,'s-',color = 'brown', label="Drama")

#Add the legend 
plt.legend()

#Add label for the x-axis
plt.xlabel("year", fontsize=15, labelpad = 15)

#Add label for the y-axis
plt.ylabel("Total revenue of the movie", fontsize=15, labelpad = 20)

#Add the plot title
plt.title("The total revenue of each genre from 2007-2016", fontsize=15, x=0.5, y=1.03)
```

##### According to the result above, Action movies might be the best generator to pick for Steven Spielberg.

### (2) The subset of director

#####  After we decide which type of movie is great for Steven Spielberg, we need to find a director who is excellent at filming this kind of movie
##### We would select those directors which have directed an action movie in the past 10 years first then compare the mean revenue of the movie they have directed

```python
##Select the columns of director,gross,and year
dire=data1[['director',"gross","genre"]]

#Select those directors which have directed Action movie in the past 10 years
dire=dire.loc[dire["genre"] =="Action"]

#Calculate the mean revenue of the movie they have directed
dire=dire.groupby('director',as_index=False).mean()

#Sort the data by gross to find the action movie director with the highest mean revenue of the movie
dire=dire.sort_values(by=['gross'],ascending=False)

#Print out the first 10 row of the result
dire.head(10)
```

##### According to the table above, James Cameron seems to be the best director to direct this movie

### (2) The visualization of director

##### Draw the bar chart of each action movie director in order to see which of them have the highest average revenue of the movies they have directed

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

#Select the column of homicides_percapita and called it as homicides
objects1 = dire.head(10)

#Select the column of director and called it as objects
objects=objects1['director']
y_pos = np.arange(len(objects))

#Select the column of gross and called it as performance
performance = objects1['gross']

#Plot the bar plot
plt.barh(y_pos, performance, align='center', alpha=0.5)

#Set locations
plt.yticks(y_pos, objects)

#Add label for the x-axis
plt.xlabel('Average revenue of the movies they have directed')

#Add the title on the plot
plt.title('Average revenue of movie of each action movie directors')

#Print out the plot
plt.show()
```

##### James Cameron seems to be the best director for Steven Spielberg

### (3) The subset of Star

##### After we decide on the director, we need to find a star who is excellent at filming an action movie
##### We would select those stars which have acted in an action movie in the past 10 years first then compare the mean revenue of the movie they have acted

```python
#Select the column of star, gross and genre
star=data1[['star',"gross","genre"]]

#Select those stars which have acted in Action movie in the past 10 years
star=star.loc[star["genre"] =="Action"]

#Calculate the mean revenue of the movie they have acted
star=star.groupby('star',as_index=False).mean()

#Sort the data by gross to find the star with the highest mean revenue of the movie they have acted
star=star.sort_values(by=['gross'],ascending=False)

#Print out the 15 stars which have the highest mean revenue they have acted
star15=star.head(15)
star15
```

##### Daisy Ridley seems to be the best to cast for the movie

### (3) The visualization of Star

##### Draw the lollipop chart of each action movie star in order to see which of them have the highest average revenue of the movies they have acted

```python
#Set the plot size
fig, ax = plt.subplots(figsize=(16,10), dpi= 80)

#Put the line on the plot
ax.vlines(x=star15.star, ymin=0, ymax=star15.gross, color='firebrick', alpha=0.7, linewidth=2)

#Put the scatter on the plot
ax.scatter(x=star15.star, y=star15.gross, s=75, color='firebrick', alpha=0.7)

#Set the plot title
ax.set_title('Average revenue of action movie of each star', fontdict={'size':22})

#Set the label of y-axis
ax.set_ylabel('Average revenue of the movies they have acted')

#Set the data it use on x-axis
ax.set_xticks(star15.star)

#Set the label of x-axis
ax.set_xticklabels(star15.star.str.upper(), rotation=60, fontdict={'horizontalalignment': 'right', 'size':12})

#Set the ylim
ax.set_ylim(0, 990000000)
```

##### Daisy Ridley seems to be the best actor for Steven Spielberg

### (4) The subset of rating

##### Decide the rating of a movie is also important so we would compare each action movie rating by both gross and score

```python
#Select the column of rating,score, gross and genre
rat=data1[['rating',"score","gross","genre"]]

#Select those rating of the movie which have been set on Action movie in the past 10 years
rat=rat.loc[rat["genre"] =="Action"]

#Calculate the mean revenue of the action movie on each rating
rat1=rat.groupby('rating',as_index=False).mean()

#Sort the data by gross and score
rat1=rat1.sort_values(by=['gross','score'],ascending=False)

#Print out the result
rat1
```

##### Set the movie on the PG-13 rating seems to be the best for Steven Spielberg

### (4) The visualization of rating

##### Draw the scatter plot of the action movie on each movie rating

```python
#Set the plot size
fig=plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')

#Select those rating of the action movie which is PG-13
c1=rat.loc[(rat["rating"] =="PG-13")]

#Select those rating of the action movie which is PG
c2=rat.loc[(rat["rating"] =="PG")]

#Select those rating of the action movie which is R
c3=rat.loc[(rat["rating"] =="R")]

#Select those rating of the action movie which is Not specified
c4=rat.loc[(rat["rating"] =="Not specified")]

#Select those rating of the action movie which is UNRATED
c5=rat.loc[(rat["rating"] =="UNRATED")]

#Select those rating of the action movie which is NOT RATED
c6=rat.loc[(rat["rating"] =="NOT RATED")]

#Set the length and the label of x and y axes
plt.gca().set(xlim=(1.0,10.0), ylim=(0,700000000),
              xlabel='score', ylabel='gross')

#Put the dot of PG-13 on the plot 
plt.scatter('score', 'gross', data=c1, s=30, c='red', label="PG-13")

#Put the dot of PG on the plot 
plt.scatter('score', 'gross', data=c2, s=30, c='blue', label="PG")

#Put the dot of R on the plot 
plt.scatter('score', 'gross', data=c3, s=30, c='green', label="R")

#Put the dot of Not specified on the plot 
plt.scatter('score', 'gross', data=c4, s=30, c='orange', label="Not specified")

#Put the dot of UNRATED on the plot 
plt.scatter('score', 'gross', data=c5, s=30, c='lightblue', label="UNRATED")

#Put the dot of NOT RATED on the plot 
plt.scatter('score', 'gross', data=c6, s=30, c='pink', label="NOT RATED")

#Add the title to the plot 
plt.title("The score and revenue of the action movie on each movie rating")

#Print out the plot
plt.legend()
```

##### According to the chart, it could be clarified to see that the action movie on the rating of PG-13 have the best outcome on both revenue and the scores

### (5) The subset of writer

##### After we decide on the director, actor, and the rating of the movie,  we still need to find a writer who is excellent at write script for the action movie
##### We would select those writers which have written script for an action movie in the past 10 years first then compare the IMDb user rating of the movie they have written

```python
#Select the column of writer,score, gross and genre
wri1=data1[['writer',"score","gross","genre"]]

#Select those writers who have written an Action movie in the past 10 years
wri1=wri1.loc[wri1["genre"] =="Action"]

#Calculate the mean revenue of the movie they have written
wri=wri1.groupby('writer',as_index=False).mean()

#Sort the data by score and gross
wri=wri.sort_values(by=['score',"gross"],ascending=False)

#Print out the 5 writer which have the highest mean score and revenue they have written
wri5=wri.head(5)
wri5
```

#### Christopher Nolan seems to be the best writer for Steven Spielberg

### (5) The visualization of writer

##### Draw the dot plot for 5 of the writer with the highest score

```python
import matplotlib.pyplot as pp

#Select the gross column in wri5 and named it as datas
datas=wri5["score"]

#Select the writer column in wri5 and named it as keys
keys=wri5["writer"]

#draw the plot with keys on x-axis and datas on y-axis
pp.plot(keys, np.zeros_like(keys) + datas, 'x', markersize=10)

#Set the size of word on x-axis
plt.xticks(size = 8)

#Add the title to the plot
plt.title('The average score 5 of the writer with the highest score')

#Print out the plot
pp.show()
```

##### According to the chart above, Christopher Nolan seems to be the best writer to write the script for this movie
